```python
import gradio as gr

with gr.Blocks() as chatbot_ui:
    gr.Markdown("## ðŸ¤– Talk to Aura ")

    chatbox = gr.Textbox(lines=2, placeholder="Ask me anything!", label="You")
    output = gr.Textbox(label="Aura")

    with gr.Row():
        submit_btn = gr.Button("Submit")
        clear_btn = gr.Button("Clear")

    submit_btn.click(fn=hf_chatbot, inputs=chatbox, outputs=output)
    chatbox.submit(fn=hf_chatbot, inputs=chatbox, outputs=output)

    def clear_inputs():
        return "", ""

    clear_btn.click(fn=clear_inputs, outputs=[chatbox, output])

chatbot_ui.launch()
```

```python
 def hf_chatbot(user_input):
    prompt = f"Answer the question: {user_input}"
    input_ids = tokenizer.encode(prompt, return_tensors="pt").to(device)

    outputs = model.generate(
        input_ids,
        max_new_tokens=100,
        do_sample=True,
        temperature=0.7,
        top_p=0.9,
        top_k=50
    )

    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return response.strip()

```

```python
from transformers import T5Tokenizer, T5ForConditionalGeneration
import torch

model_name = "google/flan-t5-base"

tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

```
